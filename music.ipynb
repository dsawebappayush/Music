```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import re

def contains_non_english_chars(text):
    pattern = re.compile(r'[^\x00-\x7F]')
    return bool(pattern.search(text))
```

```python
#taking only needed parameters
col_to_m=['popularity','duration_ms','danceability','energy','key',	'loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','time_signature']
new_df=df[col_to_m]
new_df
```

```python
#normalize
for i in new_df.columns:
    new_df[i]=new_df[i].astype(float)
    m=new_df[i].min()
    m1=new_df[i].max()
    new_df[i]=(new_df[i]-m)/(m1-m)
new_df
```

```python
#cosine simmilarity
from sklearn.metrics.pairwise import cosine_similarity

cps=cosine_similarity(new_df)
```

```python
cps
```

```python
def recommend(song):
    song_index=new_df[df['track_name']==song].index[0]
    dis=cps[song_index]
    m_l=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[2:40]
    return m_l
    
```

```python
l=recommend('Shape of You')
name=[]
for i in l:
 name.append(df.iloc[i[0]].track_name)
name
s_n=[]
m=len(name)
for i in range(0,m):
    flag=0
    for j in range(i+1,m):
        if name[i]==name[j]:
            flag=1
    if flag==0:
        s_n.append(name[i])
s_n
```
